##2. 연속형 변수 평균, 분산, 범주형 변수 빈도 계산

평균, 분산

```python
#기대수명 평균, 분산
print('기대수명 평균 : ', df['Life expectancy'].mean())
print('기대수명 분산 : ', df['Life expectancy'].var())
```

```python
#성인 사망률 평균, 분산
print('성인 사망률 평균 : ', df['Adult Mortality'].mean())
print('성인 사망률 분산 : ', df['Adult Mortality'].var())
```

```python
#영아 사망률 평균, 분산
print('영아 사망자 평균 : ', df['infant deaths'].mean())
print('영아 사망자 분산 : ', df['infant deaths'].var())
```

```python
#BMI 지수 평균, 분산
print('BMI 지수 평균 : ', df['BMI'].mean())
print('BMI 지수 분산 : ', df['BMI'].var())
```

```python
#GDP 지수 평균, 분산
print('GDP 지수 평균 : ', df['GDP'].mean())
print('GPD 지수 분산 : ', df['GDP'].var())
```

```python
#Alcohol 평균, 분산
print('알코올 평균 : ', df['Alcohol'].mean())
print('알코올 분산 : ', df['Alcohol'].var())
```

```python
#B형 간염 백신 접종률(1세 아동) 평균, 분산
print('B형 간염 백신 접종 평균 : ', df['Hepatitis B'].mean())
print('B형 간염 백신 접종 분산 : ', df['Hepatitis B'].var())
```

```python
#1인당 건강 관련 지출 GDP 대비 비율로 환산
print('1인당 건강 관련 지출 평균 : ', df['percentage expenditure'].mean())
print('1인당 건강 관련 지출 분산 : ', df['percentage expenditure'].var())
```

```python
#홍역 보고 건수(연간)
print('홍역 보고 건수 평균 : ', df['Measles'].mean())
print('홍역 보고 건수 분산 : ', df['Measles'].var())
```

```python
#5세 미만 사망자 수
print('5세 미만 사망자 평균 : ', df['under-five deaths'].mean())
print('5세 미만 사망자 분산 : ', df['under-five deaths'].var())
```

```python
#소아마비 백신 접종률
print('소아마비 백신 접종 평균 : ', df['Polio'].mean())
print('소아마비 백신 접종 분산 : ', df['Polio'].var())
```

```python
#GDP 대비 건강 비용 지출 비율
print('GDP 대비 건강 비용 지출 비율 평균 : ', df['Total expenditure'].mean())
print('GDP 대비 건강 비용 지출 비율 분산 : ', df['Total expenditure'].var())
```

```python
#디프테리아 백신 접종률
print('건강 지출 비율 평균 : ', df['Diphtheria'].mean())
print('건강 지출 비율 분산 : ', df['Diphtheria'].var())
```

```python
#HIV/AIDS로 인한 사망률(1,000명당)
print('건강 지출 비율 평균 : ', df['HIV/AIDS'].mean())
print('건강 지출 비율 분산 : ', df['HIV/AIDS'].var())
```

```python
#전체 인구
print('인구 평균 : ', df['Population'].mean())
print('인구 분산 : ', df['Population'].var())
```

```python
#1~19세 저체중 비율(퍼센트)
print('1~19세 저제충 비율 평균 : ', df['thinness 1-19 years'].mean())
print('1~19세 저체중 비율 분산 : ', df['thinness 1-19 years'].var())
```

```python
#5~9세 저체중 비율(퍼센트)
print('5~9세 저제충 비율 평균 : ', df['thinness 5-9 years'].mean())
print('5~9세 저체중 비율 분산 : ', df['thinness 5-9 years'].var())
```

```python
#소득 자원 구성 지수(0~1, 인적자본 및 소득 불평등 반영)
print('소득 자원 구성 지수 평균 : ', df['Income composition of resources'].mean())
print('소득 자원 구성 지수 분산 : ', df['Income composition of resources'].var())
```

```python
#평균 교육 년수
print('교육 년수 평균 : ', df['Schooling'].mean())
print('교육 년수 분산 : ', df['Schooling'].var())
```
